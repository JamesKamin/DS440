```python
#Load Packages

import pandas as pd

import glob

```

```python

glob.glob('/Users/jkenglish/Desktop/StormEventsData/StormEvents_details*.csv')
```

```python
#Concatenate data

dfs_combined = []

for i in glob.glob('/Users/jkenglish/Desktop/StormEventsData/StormEvents_details*.csv'):
    
    final_df = pd.read_csv(i,usecols = ['MONTH_NAME','YEAR','CZ_TIMEZONE','STATE','CZ_NAME','EVENT_TYPE','DAMAGE_CROPS','DAMAGE_PROPERTY'])
    
    dfs_combined.append(final_df)



```

```python
len(dfs_combined)
```

```python
df_concatenated = pd.concat(dfs_combined)
```

```python
df_concatenated.shape
```

```python
df_concatenated
```

```python
df_concatenated.rename(columns = {'CZ_NAME':'COUNTY_NAME'},inplace = True)

df_concatenated.rename(columns = {'CZ_TIMEZONE':'COUNTY_TIMEZONE'},inplace = True)
```

```python
df_concatenated
```

```python
df_concatenated.dtypes 
```

```python
#Convert K, M, and B to numerical form in damage_property and damage_crops columns

def val_float(value):
    if type(value) == float or type(value) == int:
        return value
    elif 'K' in value:
        if len(value) > 1:
            return float(value.replace('K', '')) * 1000
        return 1000.0
    elif 'M' in value:
        if len(value) > 1:
            return float(value.replace('M', '')) * 1000000
        return 1000000.0
    elif 'B' in value:
        return float(value.replace('B', '')) * 1000000000
    return 0.0


df_concatenated['DAMAGE_PROPERTY'] = df_concatenated['DAMAGE_PROPERTY'].apply(val_float)

df_concatenated['DAMAGE_CROPS'] = df_concatenated['DAMAGE_CROPS'].apply(val_float)
```

```python

#Drop Nan values

df = df_concatenated.dropna(subset=['DAMAGE_PROPERTY', 'DAMAGE_CROPS','COUNTY_NAME'])

df
```
